# Survival of Passengers of the Titanic

```python
Location = r'C:\Users\jwhit\jupyter\titanic.csv'
titanic = pd.read_csv(Location)
titanic
```

## Cleaning up data to solve querys

### Survivers Male Vs. Female

```python
titanic_clean = titanic[['Sex', 'Survived']]
titanic_clean
```

```python
survived_totals = titanic_clean.groupby(['Sex','Survived'])
survived_totals.size()
```

#### By grouping the data by sex we can see above that 308 Females survived Vs. 142 Males

## What class had the most survivers

```python
titanic_class = titanic[['PClass', 'Survived']]
titanic_class
```

```python
valuelist = [1]
class_survivors = titanic_class[titanic_class.Survived.isin(valuelist)]
class_survivors
```

```python
class_survivors_totals = class_survivors.groupby(['PClass','Survived'])
class_survivors_totals.size()
```

#### By removing any survivor data that did not equal 1 and grouping by PClass, we see that 1st class had the most survivors

# Polar Bear Video Data

```python
Location = r'C:\Users\jwhit\jupyter\polarbear.csv'
pbear = pd.read_csv(Location)
pbear
```

### Cleaning up the data

```python
behavior = pbear[['Bear', 'Behavior']]
behavior
```

```python
behavior['Bear'].value_counts()
```

#### From the query above we can see how many times data was recorded for each bear

```python
valuelist = ['walking']
pbear_walking = behavior[behavior.Behavior.isin(valuelist)]
pbear_walking_count = pbear_walking.groupby(['Bear', 'Behavior'])
pbear_walking_count.size()
```

#### From the query above we can see that Bear 8 was recorded walking the most and Bear 2 was recorded walking the least

```python
valuelist = ['4/21/2016 3:40:12']
pbear_time = pbear[pbear.Datetime_behavior_starts.isin(valuelist)]
pbear_time
```
